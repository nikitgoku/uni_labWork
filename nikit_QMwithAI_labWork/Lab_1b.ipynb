### Adding rows and columns

```python
# Try to add a column assign to a name that doesn't exist yet
rand_df["pigs"] = [0, 1, 1]
rand_df
```

```python
# adding rows is annoying. Try not to but if you have to...
# make your new data into a DataFrame
new_row = pd.DataFrame(data=[[1, 2, 3, 4, 5, 6, 7]], columns=rand_df.columns, index=["sun"])
new_row
```

```python
# use slices and concat to put it in the right place
inserted_df = pd.concat([rand_df.loc[:"river",], new_row, rand_df.loc["sea":,]])  # returns a new dataframe
inserted_df
```

### Deleting rows and columns

```python
# 'drop' can be used to delete both columns and rows
# first argument is name of column or index
# axis specifies if row or column: 0=row, 1=column
# inplace specifies whether to change this dataframe (True) or create a new one (False)
inserted_df.drop('pigs', axis=1, inplace=True)
inserted_df
```

```python
pruned_df = inserted_df.drop('sun', axis=0, inplace=False)
pruned_df
```

```python
# can also remove duplicate rows
# create a dataframe with duplicates
data = np.array([['', 'A', 'B', 'C'],
                 ['x', 1, 2, 3],
                 ['y', 5, 2, 1],
                 ['z', 23, 67, 1]])
duplicate_df = pd.DataFrame(data=data[1:,1:], columns=data[0,1:], index=data[1:,0])
duplicate_df
```

```python
duplicate_df.drop_duplicates(["B"], keep="last")
```

```python
# We did not specify 'inplace' so the dataframe itself remains as was...
duplicate_df
```

### Renaming rows and columns

```python
old_col_names = pruned_df.columns
old_col_names
```

```python
map_new_col_names = {old_name: new_name for old_name, new_name in zip(old_col_names, range(6))}
map_new_col_names
```

```python
pruned_df.rename(columns=map_new_col_names)  # returns new dataframe
```

### Replacing values

```python
pruned_df[pruned_df > 50] = 50
pruned_df
```

```python
pruned_df[pruned_df < 15] = 10
pruned_df
```

```python
pruned_df.replace([50, 10], ["true", "false"])
```
